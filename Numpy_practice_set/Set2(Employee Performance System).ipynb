```python
'''Set 2 (Employee Performance System)
NumPy (Intermediate)
 An HR system stores employee performance ratings (scale 1–5) for 100 employees over 4 quarters in a NumPy array.
Task:
Normalize the ratings using min-max normalization


Calculate the average rating per employee


Identify employees whose average rating is above the company mean
'''
```

```python
import numpy as np
ratings = np.random.randint(1,6,size =(100,4))
print(ratings)
```

```python
min = ratings.min()
print(min)
max = ratings.max()
print(max)
```

```python
normalize = (ratings-min)/(max-min)
print(normalize)
```

```python
avg_per_emp = ratings.mean(axis=1)
print(avg_per_emp)
```

```python
avg_mean = avg_per_emp.mean()
print(avg_mean)
```

```python
Employees = np.where(avg_per_emp >avg_mean)[0].tolist()
print(Employees)
```
