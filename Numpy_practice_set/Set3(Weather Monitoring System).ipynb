```python
'''Set 3 (Weather Monitoring System)
NumPy (Intermediate)
 A weather station stores hourly temperature readings for 7 days in a NumPy array of shape (7, 24).
Task:
Find the daily maximum and minimum temperature


Identify the day with the largest temperature variation


Replace outliers beyond Â±2 standard deviations with the mean temperature
'''
```

```python
import numpy  as np
temps = np.random.randint(0, 20 ,size = (7,24))
print(temps)
```

```python
print(temps[0])
print(temps[1])
```

```python
max = temps.max(axis =1)
print(max)
min = temps.min(axis =1)
print(min)
```

```python
variation = max -min
day_largest_variation = np.argmax(variation)
print(variation)
print(day_largest_variation)
```

```python
mean = temps.mean()
print(mean)
std = temps.std()
print(std)
```

```python
lower = mean-2*std
upper = mean=2*std
print(upper)
print(lower)
```
