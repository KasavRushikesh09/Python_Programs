```python
'''Set 4 (Online Learning Platform)
NumPy (Intermediate)
 An online platform tracks time spent (in minutes) by students on 6 courses over 14 days using a NumPy array.
Task:
Compute total time spent per course


Identify courses where average daily engagement exceeds a threshold


Rank courses based on engagement
'''
```

```python
import numpy as np
time_spent = np.random.randint(0,61,size=(14,6))
print(time_spent)
```

```python
total_time_spent_per_course = time_spent.sum(axis=0)
print(total_time_spent_per_course)
```

```python
threshold = 25
avg_daily_eng = time_spent.mean(axis = 0)
print(avg_daily_eng)
```

```python
high_course = np.where(avg_daily_eng > threshold)[0]+1
print(high_course)
```

```python
ranked_courses = np.argsort(total_time_spent_per_course)[::-1]+1
print(ranked_courses)
```
