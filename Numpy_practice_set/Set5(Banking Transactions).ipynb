```python
'''Set 5 (Banking Transactions)
NumPy (Intermediate)
 A bank stores transaction amounts for 50 accounts across 20 days in a NumPy array.

Task:
Separate debit and credit transactions


Compute daily net balance changes


Identify accounts with continuous negative balance trends'''
```

```python
import numpy as np
transaction = np.random.randint(-3000,9000,size=(20,50))
print(transaction)
print(transaction.shape)
```

```python
debit = np.where(transaction <0, transaction,0)
print(debit)
print("--------------------------------------------------------------------------------------------------------------------")
credit = np.where(transaction >0,transaction,0)
print(credit)
```

```python
daily_net_balance = transaction.sum(axis =0)
print(daily_net_balance)
```

```python
con_negeative = np.where(np.all(transaction<0,axis=1))[0]
print(con_negeative)
```

```python
print("Account with continues negative trend:",con_negeative)
print("Account with continues negative trend:",len(con_negeative))
```
