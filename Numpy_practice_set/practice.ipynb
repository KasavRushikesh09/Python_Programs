```python
'''Set 6 (Healthcare Data Analysis)
NumPy (Intermediate)
 Patient vitals (heart rate) are stored for 60 patients over 24 hours in a NumPy array.
Task:
Compute hourly average heart rate


Detect abnormal readings outside the normal range


Replace missing values using forward fill logic
'''
```

```python
import numpy as np
vitals = np.random.randint(60,120, size = (60,24))
print(vitals)
```

```python
hour_avg = vitals.mean(axis =0)
print(hour_avg)
```

```python
low ,high = 60,100
abnormal_mask = (vitals<low) | (vitals > high)
abnormal_positions = np.where(abnormal_mask)
print("Total abnormal readings:",abnormal_mask.sum())
print("Patient index:",abnormal_positions[0])
print("Hour index:",abnormal_positions[1])
```
