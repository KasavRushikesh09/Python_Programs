```python
import pandas as pd
data ={
    'emp_id': [101, 102, 103, 101, 102, 104],
    'department': ['Sales', 'IT', 'HR', 'Sales', 'IT', 'Marketing'],
    'quarter': ['Q1', 'Q1', 'Q1', 'Q2', 'Q2', 'Q2'],
    'performance_score': [85, 90, 78, 88, 92, 81]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
avg = df.pivot_table(index='department',
columns = 'quarter',
values = 'performance_score',
aggfunc = "mean")
avg
```
