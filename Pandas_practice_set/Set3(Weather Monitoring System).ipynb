```python
import pandas as pd
import numpy as np
data ={
     "date": [
        "2025-01-01", "2025-01-01", "2025-01-02", "2025-01-02",
        "2025-01-03", "2025-01-03", "2025-01-04", "2025-01-04"
    ],
    "city": [
        "Mumbai", "Delhi", "Mumbai", "Delhi",
        "Mumbai", "Delhi", "Mumbai", "Delhi"
    ],
    "temperature": [30, 25, 31, 26, 29, 24, 32, 27],
    "humidity": [70, 60, 72, 58, 68, 62, 75, 59]
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
city = df[df["city"] == 'Delhi']
print(city)
```

```python
city2 = df[df["city"] == 'Mumbai']
print(city2)
```

```python
df = df.sort_values(["city", "date"])
df["3day_avg"] = (df.groupby("city")["temperature"].transform(lambda s:s.rolling(3,min_periods=1).mean()))
```

```python
df["3day_avg"]
```

```python
df = df.sort_values(["city","date"])
df["humidity_increased"] = df.groupby("city")["humidity"].diff().gt(0)
df["humidity_increased"]
```

```python
df.loc[df["humidity_increased"],"humidity"]
```
