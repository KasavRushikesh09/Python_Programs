```python
data3 = {
    'department':['d1','d2','d1','d4','d2'],
    'salary':[1000,2000,5000,3030,6000]
}
df3 = pd.DataFrame(data3)
df3
```

```python
df3['avg_salary'] = df3.groupby('department')['salary'].transform('mean')
df3
```

```python
plt.bar(df3['department'],df3['salary'],color='red')
plt.title("DEPARTMENT  VS  SALARY")
plt.xlabel('department')
plt.ylabel('salary')
plt.show()
```

```python
df7 = pd.DataFrame({
     'date':pd.date_range(end=pd.Timestamp.today(),periods=10),
    'temp': np.random.uniform(0.0,20.0,10)
})
df7
```

```python
df7['date'] = pd.to_datetime(df7['date'])
df7
```

```python
plt.plot(df7['date'],df7['temp'],color = 'yellow')
plt.xlabel('dates')
plt.ylabel("temp")
plt.title("Temperature over last 10 days")
plt.xticks(rotation = 45)
plt.show()
```

```python
df8 = pd.DataFrame({
    'month':['jan','feb','mar','apr','may','jun'],
    'revenue':np.random.randint(20000,50000,6),
    'expenses':np.random.randint(10000,20000,6)
})
df8
```

```python
plt.plot(df8['month'],df8['revenue'],marker ='o',label="Revenue")
plt.plot(df8['month'],df8['expenses'],marker ='o',label="expenses")
plt.title("Monthly turnover")
plt.xlabel("Months")
plt.ylabel("amount")
plt.legend()
plt.show()
```

```python
df11 = pd.DataFrame({
    'day':['mon','tue','wed','thus','fri','sat','sun'],
    'daily_sales':np.random.randint(2000,5000,7)
})
df11
```

```python
plt.plot(df11['day'],df11['daily_sales'],marker= 'o', color ="red")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales Vs Days")
plt.show()
```

```python
df11['cum_sales'] = df11['daily_sales'].cumsum()
df11['cum_sales']
```

```python
plt.plot(df11['day'],df11['cum_sales'],marker= 'o', color ="red")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales Vs Days")
plt.show()
```
