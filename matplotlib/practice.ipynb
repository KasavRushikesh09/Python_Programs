```python
data = {
    "Salary":[25000,30000,37000,28000,39000,48000,55000,52000,35000,31000,26000,27000]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.shape
```

```python
# Line Plot
plt.plot(df["Salary"],color = "red",label = "salary",
         linestyle="dotted",linewidth = 2,marker = "o")
plt.show()
```

```python
#Histogram
plt.hist(df["Salary"], color ="grey",bins = 10)
plt.show()
```

```python
#Box plot
plt.boxplot(df["Salary"])
plt.show()
```

```python
#univerate: categorial
df["dept"] = ['HR','IT','Finance','HR','Finance','HR','IT','Finance','HR',"Finance","Finance","HR"]
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
#Pie chart
count = df["dept"].value_counts()
count
```

```python
plt.pie(count,labels=count.index,autopct="%1.1f",explode = [0,0.1,0.1])

plt.show()
```

```python
#Countplot
plt.bar(count.index,count , color =["orange","white","green"])
plt.show()
```

```python
df["Age"] = [22,25,29,24,30,35,40,36,28,22,23,25]
```

```python
df.head()
```

```python
plt.scatter(df["Age"],df["Salary"], color = "orange")
plt.show()
```

```python
sort_age = df.sort_values("Age")
```

```python
plt.plot(sort_age["Salary"],df["Age"],color = "red",marker = "*",linestyle = "dotted",linewidth = 2)
plt.grid()
plt.show()
```

```python
plt.bar(sort_age["Age"],df["Salary"],color = "Green",align="center")
plt.show()
```

```python
hr_salary = df[df["dept"] == "HR"]["Salary"]
hr_salary
```

```python
it_salary = df[df["dept"] == "IT"]["Salary"]
it_salary
```

```python
finance_salary = df[df["dept"] == "Finance"]["Salary"]
finance_salary
```

```python
plt.boxplot([hr_salary,it_salary,finance_salary], tick_labels=["HR","IT","Finance"])
plt.show()
```

```python
df
```

```python
salary_by_dept = df.groupby(['dept'])['Salary'].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept,labels= salary_by_dept.index,autopct = "%1.2f", explode = [0,0.1,0.1])
plt.axis("equal")
(plt.show())
```

```python
hr_mean = sum(hr_salary)/len(hr_salary)
it_mean = sum(it_salary)/len(it_salary)
finance_mean = sum(finance_salary)/len(finance_salary)
```

```python
plt.bar(["HR","IT","Finance"],[hr_mean,it_mean,finance_mean],color = ["blue","purple","skyblue"])
plt.show()
```
