```python
#Multivarient Analysis:3 numerical columns
df.head()
```

```python
df["experience"] = [1,2,3,1.5,6,9,7,8,4,1.9,4.6,6]
```

```python
df["experience"]
```

```python
df
```

```python
#Bubble plot:
plt.scatter(df['Age'],df['Salary'],s=df["experience"]*20, color="red", edgecolor= "yellow")
plt.title("Age vs Salary vs Experience")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
#2 numerical and 1 categorical column
dept_cat = df["dept"].astype("category")
dept_code = dept_cat.cat.codes
plt.scatter(df["Age"],df["Salary"],c=dept_code,cmap="rainbow")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary vs Age vs Dept")
plt.colorbar(label="Dept codes")
plt.show()
```

```python
color = {"HR":"yellow","IT":"blue","Finance":"green"}
for dept,color in color.items():
    df_dept = df[df["dept"] == dept]
    plt.scatter(df_dept['Age'],df_dept['Salary'],c =color,label=dept)
    plt.legend()
plt.colorbar(label="Dept codes")
```

```python
# Object Oriented API


fig, axs = plt.subplots(1,3, figsize = (15,5))
#Line plot
axs[0].plot(sort_age["Age"],df["Salary"],color = "red",marker = "*",linewidth = "2",markersize=2)
axs[0].grid()
axs[0].set_title("Line plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")

#histogram
axs[1].hist(df["Salary"], bins =5,color = "green")
axs[1].set_title("Histogram")
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Frequency")

#box plot
axs[2].boxplot(df["Salary"])
axs[2].set_title("boxplot")
axs[2].set_xlabel("Salary")

plt.savefig("C:\Photoes\multipleplots.png")
plt.show()
```

```python
data2 = {
    'Year':[2020,2021,2022,2023],
    'Sales':[100,150,200,250],
    'Profit':[20,30,40,50],
    'Expenses':[80,120,160,200]
}
```

```python
df2 = pd.DataFrame(data2)
```

```python
plt.plot(df2['Year'],df2['Sales'],label = "Sales", color ="yellow")
plt.plot(df2['Year'],df2['Profit'],label = "Profit", color ="red")
plt.plot(df2['Year'],df2['Expenses'],label = 'Expenses')

plt.title("Financial Analysis")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.legend()
plt.show()
```
